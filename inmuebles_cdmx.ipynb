##**Demandas de ML**

### Cuales son los valores promedio de alquiler por tipo de propiedad?

```python
datos.groupby('Tipo').mean(numeric_only=True)
```

```python
datos.groupby('Tipo')['Valor'].mean(numeric_only=True)
```

```python
datos.groupby('Tipo')[['Valor']].mean(numeric_only=True).sort_values('Valor')
```

```python
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean(numeric_only=True).sort_values('Valor')
df_tipo_precio.plot(kind='barh', figsize=(12,8),color='skyblue')
```

```python
datos.Tipo.unique()
```

```python
inmuebles_comerciales = ['Conjunto Comercial/Sala','Edificio Completo', 'Tienda/Salón', 'Casa Comercial', 'Terreno Estándar', 'Cochera/Estacionamiento', 'Galpón/Depósito/Almacén', 'Tienda en Centro Comercial', 'Hotel', 'Loteo/Condominio',
       'Industria' ]
```

```python
datos.query('@inmuebles_comerciales in Tipo')
```

```python
datos.query('@inmuebles_comerciales not in Tipo')
```

```python
df = datos.query('@inmuebles_comerciales not in Tipo')
df.head(6)
```

```python
df_tipo_precio = df.groupby('Tipo')[['Valor']].mean(numeric_only=True).sort_values('Valor')
df_tipo_precio.plot(kind='barh', figsize=(12,8),color='skyblue')
```
