### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()

```

#1. Análisis de facturación



```python
tiendas = [tienda1, tienda2, tienda3, tienda4]
for i, tienda in enumerate(tiendas, start=1):
  total = tienda["Precio"].sum()
  print(f"Tienda {i} - Total de ventas: ${total}0")


```

```python
import matplotlib.pyplot as plt

# Datos
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
total_ventas = [1150880400.00, 1116343500.00, 1098019600.00, 1038375700.00]

# Convertimos a miles de millones para mostrar en el gráfico
ventas_miles_millones = [v / 1e9 for v in total_ventas]

# Crear gráfico de barras
plt.figure(figsize=(7, 5))
bars = plt.bar(tiendas, ventas_miles_millones, color='steelblue')

# Agregar etiquetas numéricas arriba de cada barra
for bar, valor in zip(bars, ventas_miles_millones):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f"${valor:.2f}B", ha='center', va='bottom', fontsize=10)

plt.title("Total de Ventas por Tienda")
plt.ylabel("Ventas (miles de millones $)")
plt.xlabel("Tiendas")
plt.tight_layout()
plt.show()

```

# 2. Ventas por categoría

```python
for i, tienda in enumerate(tiendas, start=1):
  print(f"\nTienda {i} - Productos vendidos por categoría:")
  conteo = tienda["Categoría del Producto"].value_counts()
  for categoria, cantidad in conteo.items():
        print(f"- {categoria}: {cantidad} ventas")
```

```python
import matplotlib.pyplot as plt

# Datos de ventas por categoría por tienda
categorias = [
    "Muebles", "Electrónicos", "Juguetes", "Electrodomésticos",
    "Deportes y diversión", "Instrumentos musicales", "Libros", "Artículos para el hogar"
]

ventas_tienda1 = [465, 448, 324, 312, 284, 182, 173, 171]
ventas_tienda2 = [442, 422, 313, 305, 275, 224, 197, 181]
ventas_tienda3 = [499, 451, 315, 278, 277, 177, 185, 177]
ventas_tienda4 = [480, 451, 338, 254, 277, 170, 187, 201]

# Crear gráfico de líneas
plt.figure(figsize=(10, 6))

plt.plot(categorias, ventas_tienda1, marker='o', label="Tienda 1")
plt.plot(categorias, ventas_tienda2, marker='o', label="Tienda 2")
plt.plot(categorias, ventas_tienda3, marker='o', label="Tienda 3")
plt.plot(categorias, ventas_tienda4, marker='o', label="Tienda 4")

plt.title("Ventas por Categoría por Tienda")
plt.xlabel("Categoría")
plt.ylabel("Cantidad de Productos Vendidos")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
for i, tienda in enumerate(tiendas, start=1):
  promedio = round(tienda["Calificación"].mean(),3)
  print(f"Tienda {i} - Calificación promedio: {promedio}")
```

# 4. Productos más y menos vendidos

```python
print("Productos más y menos vendidos:")
for i, tienda in enumerate(tiendas, start=1):
  print(f"\nTienda {i}\n")
  conteo_productos = tienda["Producto"].value_counts()
  max_ventas = conteo_productos.max()
  mas_vendidos = conteo_productos[conteo_productos == max_ventas]
  print(f"Más vendido(s) ({max_ventas} ventas):")
  for producto in mas_vendidos.index:
      print(f"- {producto}")

  ventas_min = conteo_productos.min() # Mostrar el menos vendido
  menos_vendidos = conteo_productos[conteo_productos == ventas_min]
  print(f"Menos vendido ({ventas_min} ventas)")
  for producto in menos_vendidos.index:
      print(f"- {producto}")
```

```python
# Crear gráfico por tienda
for i, tienda in enumerate(tiendas, start=1):
    conteo = tienda["Producto"].value_counts()

    # Top 5 productos más vendidos
    top_5 = conteo.head(5)

    # Bottom 5 productos menos vendidos
    bottom_5 = conteo.tail(5)

    # Concatenar top primero, luego bottom
    combined = pd.concat([top_5, bottom_5])

    # Crear colores: top (verde), bottom (rojo)
    colors = ['#6aa84f'] * len(top_5) + ['#cc6666'] * len(bottom_5)

    # Invertimos todo para que los más vendidos queden arriba (en el gráfico)
    combined = combined[::-1]
    colors = colors[::-1]

    # Graficar
    plt.figure(figsize=(10, 6))
    combined.plot(kind='barh', color=colors)
    plt.title(f"Tienda {i} - Top 5 y Bottom 5 productos vendidos")
    plt.xlabel("Cantidad de ventas")
    plt.ylabel("Producto")
    plt.tight_layout()
    plt.show()



```

# 5. Envío promedio por tienda

```python
for i, tienda in enumerate(tiendas, start=1):
  promedio = round(tienda["Costo de envío"].mean(),2)
  print(f"Tienda {i} - Costo de envío promedio: ${promedio}")
```

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
calificaciones = [3.977, 4.037, 4.048, 3.996]
costos_envio = [26018.61, 25216.24, 24805.68, 23459.46]

# Crear gráfico de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(calificaciones, costos_envio, color='mediumslateblue', s=100)

# Añadir etiquetas a cada punto
for i in range(len(tiendas)):
    plt.text(calificaciones[i] + 0.002, costos_envio[i], tiendas[i], fontsize=10)

plt.title("Calificación Promedio vs. Costo de Envío Promedio")
plt.xlabel("Calificación Promedio")
plt.ylabel("Costo de Envío Promedio ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
